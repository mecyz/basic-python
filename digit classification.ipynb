```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D,Flatten, Dense
```

```python
#train and visualise the MNIST dataset
(x_train,y_train), (x_test, y_test) = mnist.load_data()
```

```python
plt.imshow(x_train[0], cmap='gray')
plt.title(f'label: {y_train[0]}')
plt.show()
```

```python
plt.imshow(x_train[5], cmap='gray')
plt.title(f'label: {y_train[5]}')
plt.show()
```

```python
fig, axes = plt.subplots(1, 10, figsize=(15, 3))
for i in range(10):
    axes[i].imshow(x_train[i], cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(f'Label: {y_train[i]}')
plt.show()
```

```python
c=0
for i in range(len(x_train)):
  if(y_train[i]==5):
    plt.imshow(x_train[i], cmap='gray')
    plt.title((f'label: {y_train[i]}'))
    plt.show()
    c=c+1
    if c==5:
      break
```

```python
x_train=x_train.reshape((x_train.shape[0],28,28,1)).astype('float32') /255
x_test=x_test.reshape((x_test.shape[0],28,28,1)).astype('float32') /255
plt.imshow(x_train[5],cmap='gray')
plt.title(f'Label:{y_train[5]}')
plt.show()
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(6, activation='relu'),
    Dense(10, activation='softmax'),
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1, validation_split=0.2)
```

```python
test_loss, test_acc=model.evaluate(x_test,y_test)
print(f'Test accuracy: {test_acc}')
```

```python
predictions=model.predict(x_test)

plt.imshow(x_test[0].reshape(28,28), cmap='gray')
plt.title((f'Predicted: {np.argmax(predictions[0])}'))
```

```python
c=0
for i in range (len(x_test)):
   if(y_test[i] != np.argmax(predictions[i])):
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f'label pred: {np.argmax(predictions[i])}, label actu:{y_test[i]}')
    plt.show()
    c=c+1
```
